```python
import pandas as pd
df = pd.read_csv("StudentPerformance_1000.csv")
df.head(50)
```

```python
df["total_score"] = df["math score"] + df["reading score"] + df["writing score"]  #addition
df[["math score", "reading score", "writing score", "total_score"]]
```

```python
df["average_score"] = df["total_score"] / 3  #division
df[["total_score", "average_score"]].head()
```

```python
df["math_reading_diff"] = df["math score"] - df["reading score"]  #sub
df[["math score", "reading score", "math_reading_diff"]].head()
```

```python
df["math_pass"] = df["math score"] >= 40    #boolean
df[["math score", "math_pass"]].head()
```

```python
df["mul_weighted_score"] = (df["math score"] * 1.5) + (df["reading score"] * 1.5) + (df["writing score"] * 1.5) #mul
df[["math score", "reading score", "writing score", "mul_weighted_score"]].head()
```

```python
#conditional statements
def score_category(score):
    if score >= 90:
        return "Excellent"
    elif score >= 60:
        return "Good"
    else:
        return "Needs Improvement"
print(score_category(75))
```

```python
#loops
for_loop_result = []
for score in [80, 50, 30]:
    if score < 40:
        break
    for_loop_result.append(score)
print('For loop:', for_loop_result)

while_loop_result = []
i = 0
while i < 3:
    while_loop_result.append(i)
    i += 1
print('While loop:', while_loop_result)
```

```python
import pandas as pd
from functools import reduce

# Sample Data (add more rows or load from file if needed)
data = {
    "math score": [80, 37, 84, 58, 47],
    "reading score": [56, 45, 81, 90, 27],
    "writing score": [93, 45, 32, 85, 18]
}

df = pd.DataFrame(data)
# Lambda with filter: Get passing students (math score > 50)
passing = list(filter(lambda x: x > 50, df['math score']))
print("Passing Math Scores:", passing)

# Lambda with map: Increase reading score by 5
updated_reading = list(map(lambda x: x + 5, df['reading score']))
print("Updated Reading Scores:", updated_reading)

# Lambda with reduce: Total writing score
total_writing = reduce(lambda a, b: a + b, df['writing score'])
print("Total Writing Score:", total_writing)

```
